```python
X1 = df.loc[:,["Age","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans= KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(X1)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X1[:,0],X1[:,1],c=kmeans.labels_,cmap="rainbow")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="black")
plt.title("Clusters of Customers")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

inertias = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df[["Annual Income (k$)", "Spending Score (1-100)"]])
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 11), inertias, "-o")  # Plot the elbow curve
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
hierarchy = AgglomerativeClustering(n_clusters=5, linkage='ward', affinity='euclidean')
hierarchy.fit(df[["Annual Income (k$)", "Spending Score (1-100)"]])
df["hierarchy_cluster"] = hierarchy.labels_

plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c=hierarchy.labels_)
plt.title("Hierarchical Clustering")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
dbscan = DBSCAN(eps=10, min_samples=5)  # Experiment with different values
dbscan.fit(df[["Annual Income (k$)", "Spending Score (1-100)"]])
df["dbscan_cluster"] = dbscan.labels_
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c=hierarchy.labels_)
plt.title("DBSCAN Clustering")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
# Silhouette analysis
kmeans_silhouette = silhouette_score(df[["Annual Income (k$)", "Spending Score (1-100)"]], kmeans.labels_)
hierarchy_silhouette = silhouette_score(df[["Annual Income (k$)", "Spending Score (1-100)"]], hierarchy.labels_)
dbscan_silhouette = silhouette_score(df[["Annual Income (k$)", "Spending Score (1-100)"]], dbscan.labels_)

print("Silhouette Score (KMeans):", kmeans_silhouette)
print("Silhouette Score (Hierarchical):", hierarchy_silhouette)
print("Silhouette Score (DBSCAN):", dbscan_silhouette)
```
